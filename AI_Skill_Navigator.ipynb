```python
import PyPDF2

pdf_file = list(uploaded.keys())[0]

resume_text = ""

with open(pdf_file, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        text = page.extract_text()

        if text:
            resume_text += text

print("Resume Extracted Successfully")
print("\n")
print(resume_text[:1000])
```

```python
skill_database = [
    "python",
    "java",
    "c++",
    "sql",
    "mysql",
    "machine learning",
    "deep learning",
    "tensorflow",
    "pytorch",
    "flask",
    "django",
    "html",
    "css",
    "javascript",
    "react",
    "nodejs",
    "mongodb",
    "power bi",
    "excel",
    "data analysis",
    "nlp",
    "computer vision",
    "docker",
    "aws",
    "git",
    "github"
]
```

```python
resume_lower = resume_text.lower()

detected_skills = []

for skill in skill_database:

    if skill in resume_lower:
        detected_skills.append(skill)

print("Detected Skills:")
print("----------------")

for skill in detected_skills:
    print(skill)
```

```python
job_roles = {

    "Data Analyst":[
        "python",
        "sql",
        "excel",
        "power bi"
    ],

    "Machine Learning Engineer":[
        "python",
        "machine learning",
        "tensorflow",
        "docker"
    ],

    "AI Engineer":[
        "python",
        "deep learning",
        "tensorflow",
        "pytorch"
    ],

    "Backend Developer":[
        "python",
        "flask",
        "django",
        "sql"
    ],

    "Data Scientist":[
        "python",
        "sql",
        "machine learning",
        "data analysis"
    ]
}
```

```python
recommendations = []

for role, skills in job_roles.items():

    matched = len(
        set(detected_skills).intersection(set(skills))
    )

    score = (matched / len(skills)) * 100

    recommendations.append(
        (role, round(score,2))
    )

recommendations.sort(
    key=lambda x: x[1],
    reverse=True
)

print("Recommended Roles")
print("------------------")

for role, score in recommendations:
    print(role, ":", score,"%")
```

```python
best_role = recommendations[0][0]

print("Best Matching Role:")
print(best_role)
```

```python
required_skills = job_roles[best_role]

missing_skills = []

for skill in required_skills:

    if skill not in detected_skills:
        missing_skills.append(skill)

print("Missing Skills")
print("--------------")

for skill in missing_skills:
    print(skill)
```

```python
print("Personalized Learning Roadmap")
print("-----------------------------")

month = 1

for skill in missing_skills:

    print(
        f"Month {month}: Learn {skill}"
    )

    month += 1
```

```python
print("="*50)
print("AI SKILL NAVIGATOR REPORT")
print("="*50)

print("\nDetected Skills:")
for skill in detected_skills:
    print("-", skill)

print("\nRecommended Role:")
print(best_role)

print("\nMissing Skills:")
for skill in missing_skills:
    print("-", skill)

print("\nLearning Roadmap:")
month = 1

for skill in missing_skills:
    print(f"Month {month}: {skill}")
    month += 1
```

```python
ats_score = (len(detected_skills) / len(skill_database)) * 100

print("ATS Score:", round(ats_score, 2), "%")
```

```python
job_description = """
Looking for a Machine Learning Engineer with Python,
TensorFlow, Docker and Machine Learning skills.
"""

jd_lower = job_description.lower()

jd_skills = []

for skill in skill_database:
    if skill in jd_lower:
        jd_skills.append(skill)

match = len(set(detected_skills).intersection(set(jd_skills)))

match_score = (match / len(jd_skills)) * 100

print("Job Match Score:", round(match_score,2), "%")
```
